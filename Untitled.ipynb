```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
dataFile = "static/data/combined_data.csv"
```

```python
mn_ed_df = pd.read_csv(dataFile)
mn_ed_df.head()
```

```python
mn_ed_df.drop(mn_ed_df[mn_ed_df["avgteachersalary"] == 0].index, inplace = True)

mn_ed_df.head()
```

```python
salaries = mn_ed_df["avgteachersalary"]
fig, ax = plt.subplots(figsize=(6.5, 9.5))
color=('#ebdef0')
ax.set_title("Avg. MN Teacher Salary by District, 2018")
ax.set_ylabel("Salary ($/year)", fontsize=14)
bplot = ax.boxplot(salaries, showmeans=True, meanline=True, notch=True, showfliers=True)
ax.yaxis.grid(True, linestyle = '-', which='major', color='#a9cce3', alpha=0.5)
# boxprops = dict(linestyle="--", linewidth=5, color="darkgoldenrod")
plt.setp(bplot['boxes'], color="#1a0f46", linewidth=2.5)
plt.setp(bplot['whiskers'], color="purple", linewidth=2.5)
plt.setp(bplot['fliers'], color="#e80e75")
plt.show
plt.setp(bplot['medianline'], linewidth=2.5)
orange_patch = mpatches.Patch(color="orange", label= "Median Salary")
green_dash = mpatches.Patch(color="green", linestyle="-", label = "Mean Salary")
plt.legend(handles=[orange_patch, green_dash],loc='best')
```
